### Detection Face With Haarcascade_frontalface

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

**Deteksi Muka**

```python
#foto = cv2.imread('/content/drive/MyDrive/bahan/groubwajah.jpg')

foto = cv2.imread('/content/drive/MyDrive/bahan/_face_detection.jpg')
deteksimuka = cv2.CascadeClassifier('/content/drive/MyDrive/bahan/haarcascade_frontalface_default.xml')

rgb = cv2.cvtColor(foto,cv2.COLOR_BGR2RGB)
abu = cv2.cvtColor(rgb,cv2.COLOR_RGB2GRAY)
muka = deteksimuka.detectMultiScale(abu, 1.3,1)
print(muka)

for (x,y,w,h)in muka:
  cv2.rectangle(foto,(x,y),(x+w,y+h),(0,255,0),2)
  kotakabu = abu[y:y+h, x:x+w]
  kotakwarana = foto[y:y+h, x:x+w]

  plt.imshow(cv2.cvtColor(foto,cv2.COLOR_BGR2RGB))
```
